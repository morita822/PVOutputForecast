```python
import os
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
class CFG:
    #memo = "LGBM_025でbest_modelをsave"
    note_num = "StackingOpt/EDA001"
    # n_folds = 5
    # trn_fold = list(range(n_folds))
    #trn_fold = [0]
    seed = 42


OUTPUT_DIR = f'H:/study/output/{CFG.note_num}/'

if not os.path.exists(OUTPUT_DIR):
    os.makedirs(OUTPUT_DIR)
```

```python
#全てのfiles_pathを取得する
dir_path = 'H:\study\error_data\kanto_parameter_lambda_0.019_iterate_1000_p3'
files_path = []
for current_dir, sub_dirs, files_list in os.walk(dir_path): 
  for file_name in files_list: 
    #print(os.path.join(current_dir,file_name))
    files_path.append(os.path.join(current_dir,file_name))
```

```python
year13 = [s for s in files_path if 'error_13' in s]
year14 = [s for s in files_path if 'error_14' in s]
```

```python
df_list = []
usecols = ["id","twoweeks_max","flo_lng","flo_lat","flo","nv","nv_meshint","env","preal","preal_meshint","ppred","nv_max_perror","nv_meshint_max_perror"]
year = 2013
datetime_list = []
for i,file_path in enumerate(year13):

  df = pd.read_csv(file_path,encoding="shift-jis",usecols=usecols)

  s = file_path.split("\\")[-2]
  month = int(s[:s.find('月')])
  day = int(s[s.find('月')+1:s.find('日')])
  time = file_path.split("\\")[-1].split('_')[1][len(f"{year}{month}{day}")-2:]
  hour = int(float(time)//1)
  minute = int(float(time)%1*60)

  t = (year,month,day,hour,minute)
  dt = datetime(year,month,day,hour,minute)

  df["datetime"] = dt
  datetime_list.append(dt)

  df_list.append(df)

  if i%300 == 0:
    print(f"{i/len(year13)*100}%_finish")

  #break
```

```python
df_list2 = []
usecols = ["id","twoweeks_max","flo_lng","flo_lat","flo","nv","nv_meshint","env","preal","preal_meshint","ppred","nv_max_perror","nv_meshint_max_perror"]
year = 2014
datetime_list = []
for i,file_path in enumerate(year14):

  df = pd.read_csv(file_path,encoding="shift-jis",usecols=usecols)

  s = file_path.split("\\")[-2]
  month = int(s[:s.find('月')])
  day = int(s[s.find('月')+1:s.find('日')])
  time = file_path.split("\\")[-1].split('_')[1][len(f"{year}{month}{day}")-2:]
  hour = int(float(time)//1)
  minute = int(float(time)%1*60)

  t = (year,month,day,hour,minute)
  dt = datetime(year,month,day,hour,minute)
  datetime_list.append(dt)

  df["datetime"] = dt

  df_list2.append(df)

  if i%300 == 0:
    print(f"{i/len(year14)*100}%_finish")

  #break
```
